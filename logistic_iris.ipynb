```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.linear_model import LogisticRegression
logistic = LogisticRegression(random_state=1) #인자에 random_state값을 지정해주는건 모델에서 최적화 알고리즘을 쓴다는 뜻
```

```python
path = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
df = pd.read_csv(path, header=None)
```

```python
df.head()
```

```python
df.columns = ['sepal length', 'sepal width', 'petal length', 'petal width', 'species']
```

```python
df.head(2)
```

```python
X = df.drop('species', axis=1)
```

```python
y = df['species']
y
```

```python
y.value_counts()
```

```python
d = {'Iris-setosa':0, 'Iris-versicolor' :1, 'Iris-virginica':2}
```

```python
y_encoded = y.map(lambda x: d[x])
```

```python
y_encoded
```

```python
X.head(2)
```

```python
type(X)
```

```python
X = X[['sepal length', 'petal length']] # 여러개 쪼갤땐 대괄호 두번
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, train_size=0.7, random_state=1, stratify=y)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y.shape
```

```python
y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler() # 평균0 표준편차 1로 표준화
sc.fit(X_train)
```

```python
sc.transform(X_train)
```

```python
sc.transform(X_test)
```

```python
plt.show()
```

```python
plt.title("plt연습")
```

```python
X_train_std = sc.transform(X_train)
```

```python
X_test_std = sc.transform(X_test)
```

```python
type(X_train)
```

```python
type(X_train_std)
```

```python
X0 = X_train_std[y_train==0]
X1 = X_train_std[y_train==1]
X2 = X_train_std[y_train==2]
```

```python
X0
```

```python
plt.scatter(X0[:,0], X0[:,1], marker = 's', label='0: santosa')
plt.scatter(X1[:,0], X1[:,1], marker = 'x', label='1: versicolor')
plt.scatter(X2[:,0], X2[:,1], marker = 'o', label='2: virginica')
```

```python
from sklearn.linear_model import LogisticRegression

logistic = LogisticRegression(random_state = 1)
logistic.fit(X_train_std, y_train) ## std된걸로 학습시켜야지..
```

```python
logistic.predict(X_test_std[:3])
```

```python
logistic.predict(X_test_std[:3, :])
```

```python
logistic.score(X_train_std, y_train)
```

```python
logistic.score(X_test_std, y_test)
```
