# Pandas

pandas: 구조화된 데이터를 빠르고 쉽게 다양한 형식으로 가공할 수 있는 강력한 데이터 분석 환경 제공

구조화된 데이터: 행렬로 이뤄진, 엑셀데이터
- numpy를 기반으로 하는 패키지

### NumPy vs Pandas
Numpy:
 - 단일 데이터 타입
 - 행/열이 단순 정수형 인덱스

Pandas:
 - 여러 데이터 타입 허용
 - 행/열을 레이블(이름)로 식별(
  ex) 중간고사 성적)

````python
import numpy as np
import pandas as pd
````

# Pandas 객체

## DataFrame 객체
- 행과 열에 레이블이 부착된 다차원 배열
- 강력한 데이터 연산 지원

DataFrame 객체 생성하기


```
pd.DataFrame(data, index, columns, dtype, copy)
```



파이썬 리스트로 데이터프레임 만들기

````python
pd.DataFrame([[1, 2, 3], [4,5,6]], columns = ['x', 'y', 'z'], index = ['정', '김'])
````

넘파이 배열로 데이터프레임 만들기

````python
np.random.seed(0)
````

````python
scores = pd.DataFrame(np.random.randint(101, size=(5,2)),
                      columns = ['mid', 'final'], index=[1, 3, 9, 6, 5])
print(scores)
````

DataFrame의 칼럼이름과 인덱스들은 판다스의 인덱스 객체임

````python
scores.index
````
